```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

# matplotlib 한글 출력 가능하도록 만들기
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="/System/Library/AssetsV2/com_apple_MobileAsset_Font6/ee9e9584cd1a5f676e03a9efcc80783bc7e12e2c.asset/AssetData/NanumGothic.ttc").get_name()
rc('font', family=font_name)

import warnings
warnings.filterwarnings("ignore")
```

```python
# korea/usa/japan : 각 나라별 데이터
# data : 다 합친 데이터
data= pd.read_csv('data.csv', encoding='euc-kr')
korea = pd.read_csv('korea_final.csv', encoding = 'euc-kr')
usa = pd.read_csv('MLB_final.csv', encoding = 'euc-kr')
japan = pd.read_csv('NPB_final.csv', encoding = 'euc-kr')
```

```python
data = data.drop('Unnamed: 0', axis =1 )
korea = korea.drop('Unnamed: 0', axis =1 )
japan = japan.drop('Unnamed: 0', axis =1 )
usa = usa.drop('Unnamed: 0', axis =1 )
```

```python
# usa와 japan은 열이 똑같음
# 행손실없음 확인
uj = pd.concat([usa, japan], axis=0)
print(usa.shape)
print(japan.shape)
print(uj.shape)
print(usa.shape[0]+japan.shape[0])
```

```python
# 전체 데이터 행 손실 없음
# row
print(korea.shape)
print(uj.shape)
print(data.shape)
print(korea.shape[0]+uj.shape[0])
```

```python
# 전체 열 손실 없음
#columns
print(korea.shape)
print(uj.shape)
print(data.shape)
print(len(set(korea.columns | uj.columns)))
```

```python
korea.columns & usa.columns & japan.columns
```

- 위 columns 들을 제외하고는 na값이 존재 
- korea의 경우에는 총3개의 제공기관에서 오는 제공 기록 차이에 따라 발생
- h : hitter 기록 p : pitcher 기록
- 1: 제공기관1 2 : 제공기관2
- 한국자료 말고는 제공기관2가 없으므로 해당 column은 전부 NA
- 특정기관에서 제공하는 지표의 경우 해당자료 제외하고는 NA
- 변수명앞에 s가 붙은경우 statiz 자체 제공 index

#### 여기서부터는 리그별 각 column의 kdeplot 또는 boxplot이 계속 나열됩니다
#### 파이썬 오류때문에 나눠서 그렸고 한/미/일 을 비교할 수 있는 column만 채택했습니다. 여기에 포함되지 않은 columns은 각 나라별 column의 EDA를 보시면 됩니다.

#### 1. 미국의 분포가 더 높게 분포
#### 2. 별 차이 없음 
#### 보고서에는 변수 몇개의 분포만 집어 넣으면 될듯합니다.


```python
sns.kdeplot(data1[data1.League == 'KBO']['PA_h1'])
sns.kdeplot(data1[data1.League == 'MLB']['PA_h1'])
sns.kdeplot(data1[data1.League == 'NPB']['PA_h1'])


plt.title('리그별 PA_h1')
plt.legend(['KBO', 'MLB','NPB'])
```

```python
sns.kdeplot(data1[data1.League == 'KBO']['AB_h1'])
sns.kdeplot(data1[data1.League == 'MLB']['AB_h1'])
sns.kdeplot(data1[data1.League == 'NPB']['AB_h1'])


plt.title('리그별 AB_h1')
plt.legend(['KBO', 'MLB','NPB'])
```

```python
fig, axes = plt.subplots(nrows = 7, figsize=(10,50))
a = ['PA_h1', 'AB_h1', 'R_h1', 'H_h1','X2B_h1', 'X3B_h1', 'HR_h1']
for i in  range(0,len(a)) :    
    val = a[i]
    sns.boxplot(data=data1, x='League', y= val , ax=axes[i] )
    axes[i].set_title('리그별 {}'.format(val))
```
