```python
col_team_off = [['Yr', 'Team','HR','R','RBI','SB','BB%','K%','ISO','BABIP','AVG','OBP','SLG','wOBA','wRC+','EV','BSR','OFF','DEF','WAR']]
col_team_def = [['Yr', 'Team',  'W', 'L','SV', 'G', 'GS', 'IP', 'K/9', 'BB/9', 'HR/9', 'BABIP',
                 'LOB%', 'GB%', 'HR/FB', 'EV', 'ERA', 'FIP', 'xFIP', 'WAR']]

team_off = pd.DataFrame(columns = col_team_off)
team_def = pd.DataFrame(columns = col_team_def)
len(col_team_off[0])
```

```python
year = range(2020, 1981, -1)
for yr in year:
    url = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=0&type=8&season='+str(yr)+'&month=0&season1='+str(yr)+'&ind=0&team=0,ts&rost=&age=&filter=&players=0'
    driver.get(url)

    driver.implicitly_wait(15)
    table = driver.find_element_by_xpath('/html/body/form/div[3]/div[2]/span[2]/div/table/tbody').get_attribute("innerHTML") 
    soup = BeautifulSoup(table, 'html.parser') 
    for i in range(0,len(soup.findAll('tr'))):
        team_name = soup.findAll('tr')[i].get_text()[3:]
        data = soup.findAll('tr')[i]
        data = data.findAll('td')[4::]
        for j in range(0, len(data)):
            data[j] = data[j].get_text()
        data = [yr, team_name] + data
        data = pd.DataFrame(data).T
        data.columns = team_off.columns
        team_off = pd.concat([team_off, data])
for yr in year:
    url = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=all&qual=0&type=8&season='+str(yr)+'&month=0&season1='+str(yr)+'&ind=0&team=0,ts&rost=&age=&filter=&players=0'
    driver.get(url)
    table = driver.find_element_by_xpath('/html/body/form/div[3]/div[2]/span[2]/div/table/tbody').get_attribute("innerHTML") 
    soup = BeautifulSoup(table, 'html.parser') 
    for i in range(0,len(soup.findAll('tr'))):
        team_name = soup.findAll('tr')[i].get_text()[3:]
        data = soup.findAll('tr')[i]
        data = data.findAll('td')[2::]
        for j in range(0, len(data)):
            data[j] = data[j].get_text()
        data = [yr, team_name] + data
        data = pd.DataFrame(data).T
        data.columns = team_def.columns
        team_def = pd.concat([team_def, data])
```
